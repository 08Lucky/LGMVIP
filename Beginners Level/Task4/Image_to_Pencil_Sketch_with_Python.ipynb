**Importing all of the essential libraries**

```python
import cv2
from cv2_plt_imshow import cv2_plt_imshow
import matplotlib.pyplot as plt
```

**read the image:**

```python
image = cv2.imread("kitten.jpg")
cv2_plt_imshow(image)
cv2.waitKey(0)
```

**we will create a new image by converting the original image to greyscale:**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_plt_imshow( gray_image)
cv2.waitKey(0)
```

**invert the new grayscale image:**

```python
inverted_image = 255 - gray_image
cv2_plt_imshow( inverted_image)
cv2.waitKey()
```

**blur the image by using the Gaussian Function in OpenCV:**

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

**final step is to invert the blurred image, then we can easily convert the image into a pencil sketch:**

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_plt_imshow(pencil_sketch)
cv2.waitKey(0)
```

**Let's compare both the original image and the pencil sketch**

```python
img1 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
img2 =cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB)
NUM_ROWS = 1
IMGs_IN_ROW = 2
f, ax = plt.subplots(NUM_ROWS, IMGs_IN_ROW, figsize=(16,6))
ax[0].imshow(img1)
ax[1].imshow(img2)
ax[0].set_title('Original Image')
ax[1].set_title('Pencil Sketch')
plt.tight_layout()
plt.show()
```
